```python
price_eth = data_eth["Adj Close"].to_list()
def sum_of_eth(list, size):
    if (size == 0):
        return 0
    else:
        return list[size - 1] + sum_of_eth(list, size - 1)
    
sum_eth = sum_of_eth(price_eth, len(price_eth))
print(sum_eth)
mean_eth = sum_eth/len(price_eth)
print(f"Mean: {mean_eth}")
print(f"Max price: {max(price_eth)}")
print(f"Min price: {min(price_eth)}")
```

```python
price_btc = data_btc["Adj Close"].to_list()
# sum_btc = sum(price_btc)

def sumOfBTC(list, size):
   if (size == 0):
     return 0
   else:
     return list[size - 1] + sumOfBTC(list, size - 1)
  
  
sum_btc = sumOfBTC(price_btc, len(price_btc))


mean_btc = sum_btc/len(price_btc)
print(f"Sum: {sum_btc}")
print(f"Sum: {float('{:.2f}'.format(sum_btc))}")
print(f"Mean: {mean_btc}")
print(f"Max price: {max(price_btc)}")
print(f"Min price: {min(price_btc)}")
```
